```python
def congestion(row):
    if row['numbikesavailable'] == 0 and row['capacity'] !=0:
        return True
    else:
        return False
    
```

```python
df['lost_revenue'] = 1
```

```python
#df['lost_revenue']= df['lost_revenue'].apply(congestion())
df['lost_revenue'] = df.apply(lambda row : congestion(row), axis=1) 
```

```python
df.head()
```

```python

mask = df['lost_revenue'].apply(lambda x: x == False)
selected_columns = ['name','stationcode']
list_of_congested_stations = df.loc[mask,selected_columns]
```
